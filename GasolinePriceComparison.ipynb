```python
import pandas as pd
```

```python
df = pd.read_csv(r"C:\Users\Fricks Nathaniel\Documents\coba-cobaporto\world-data-2023.csv")
```

```python
df
```

```python
#Removing $ in Gasoline price 
df["Gasoline Price"] = df["Gasoline Price"].str.replace('$', '')
```

```python
df["Gasoline Price"]
```

```python
#Removing $ in Minimum wage 
df["Minimum wage"] = df["Minimum wage"].str.replace('$', '')
```

```python
df["Minimum wage"]
```

```python
columns_to_convert = ["Gasoline Price", "CPI","Minimum wage"]
```

```python
#converting Gasoline Price, CPI, and Minimum wage  data types
df[columns_to_convert] = df[columns_to_convert].apply(pd.to_numeric, errors='coerce')
```

```python
missing_values = df.isna().sum()
```

```python
print(missing_values)
```

```python
#Filling missing values in CPI using mean
df["CPI"].fillna(df["CPI"].mean(), inplace=True)
```

```python
df["CPI"]
```

```python
#Filling missing values in Gasoline Price using median
df["Gasoline Price"].fillna(df["Gasoline Price"].median(), inplace=True)
```

```python
df["Gasoline Price"]
```

```python
#Filling missing values in Minimum wage using median
df["Minimum wage"].fillna(df["Minimum wage"].median(), inplace=True)
```

```python
df["Minimum wage"]
```
